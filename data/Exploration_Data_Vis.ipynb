```python
df = pd.read_csv("topChartWithFeatures.csv")
# print(data.head())
```

```python
post_malone = df.loc[(df['Artist']=='Post Malone')]
post_malone_date = post_malone.loc[(df['Date']==1/1/18)]

print(post_malone.head())
```

```python
objects = ('Rockstar', 'I Fall Apart', 'Candy Paint', 'Congratulations', 'Go Flex')
y_pos = np.arange(len(objects))
performance = [post_malone['Tempo'][0],post_malone['Tempo'][7],post_malone['Tempo'][11],post_malone['Tempo'][22],post_malone['Tempo'][79]]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Tempo')
plt.title("Posty's Top Songs")

plt.show()
```
